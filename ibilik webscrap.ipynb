```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
# Base URL
base_url = "https://www.ibilik.my/rooms/kuala_lumpur?page={}"

# Headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# Initialize data storage
all_data = []

# Define the maximum pages to scrape
max_pages = 556  

for page in range(445, max_pages + 1):
    url = base_url.format(page)
    print(f"Scraping page {page}: {url}")

    # Send request
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        listings = soup.find_all("div", class_="home-list-pop")

        for listing in listings:
            title = listing.find("h3").text.strip() if listing.find("h3") else "No title"
            price = listing.find("div", class_="room_price").text.strip() if listing.find("div", class_="room_price") else "No price"
            location = listing.find("a", title="Room Rental in Kuala Lumpur").text.strip() if listing.find("a", title="Room Rental in Kuala Lumpur") else "No location"
            link = listing.find("a")["href"] if listing.find("a") else "No link"

            # Extract descriptions (e.g., facilities, rental conditions, etc.)
            description_tags = listing.find_all("p")
            descriptions = [desc.text.strip() for desc in description_tags]
            
            # Split into different columns (Location, Rental Conditions, Room Type, Facilities)
            description_parts = descriptions + [""] * (4 - len(descriptions))  
            location_detail, rental_conditions, room_type, facilities = description_parts[:4]

            # Append data
            all_data.append([title, price, location, link, location_detail, rental_conditions, room_type, facilities])

        # Add a random delay (1 to 3 seconds) to avoid detection
        time.sleep(random.uniform(5, 8))

    else:
        print(f"Failed to fetch page {page}. Status Code:", response.status_code)
        break  # Stop if blocked

# Convert to DataFrame and save
df = pd.DataFrame(all_data, columns=["Title", "Price", "Location", "Link", "Location Detail", "Rental Conditions", "Room Type", "Facilities"])

print("Scraping completed!")
```
